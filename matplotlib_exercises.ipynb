```python
import matplotlib.pyplot as plt
# 1. Use matplotlib to plot the following equation: 
#You'lleed to write the code that generates the x and y points.
#Add an anotation for the point 0, 0, the origin.

plt.figure(figsize = (3,3))
x = list(range(-100,101))
y = [ n ** 2 - n + 2 for n in x]
plt.title('quadratic equation')
plt.xlabel('x values')
plt.ylabel('y values')
plt.xticks(rotation=45)
plt.yticks(rotation = 46)
plt.plot(x, y, c = "green", alpha = 1,)
plt.annotate("origin", xy = (0,0), xytext = (0,3000), 
             arrowprops = {'facecolor' : "black",
                           'arrowstyle' : "<-"
                            }
             )
plt.show()
```

```python
#3. Combine the figures you created in the last step into one large figure with 4 subplots.
plt.figure(figsize = (10,10))

#y=√x
import math
x= list(range(0,100))
y = [math.sqrt(n) for n in x]
plt.subplot(2,2,1)
plt.plot(x,y, c = "red")
plt.title('sqaureroot of x')


#y=x^3
x = list(range(-10,10))
y = [n**3 for n in x]
plt.subplot(2,2,2)
plt.plot(x,y, c = "green")
plt.title('cube of x')


#y=2**x
x= list(range(-5,10))
y= [2**n for n in x]
plt.subplot(2,2,3)
plt.plot(x,y, c = "blue")
plt.title('$2^x$')


#y=1/(x+1)
x = range(0,114)
y= [1/(n+1) for n in x]
plt.subplot(2,2,4)
plt.title('$1/(x+1)$')
plt.plot(x,y, c = "yellow")

plt.suptitle('graphs')
plt.show()
```
