```python
import matplotlib.pyplot as plt
import math
```

```python
x = list(range(1,15))
x
```

```python
plt.figure(figsize = (12,8))

y1 = [(0*n + 1) for n in x]
plt.plot(x,y1,c="red",label='O(1)')


y2 = [(math.log(n)) for n in x]
plt.plot(x,y2,c="green",label='$O(log n)$')

y3 = [n for n in x]
plt.plot(x,y3, c='yellow', label = "n")

y4 = [n * math.log(n) for n in x]
plt.plot(x,y4,c="blue",label="$O(n logn)$")

y5 = [n * n for n in x]
plt.plot(x,y5,c='gray',label='$O^2$')

y6 = [2**n for n in x]
plt.plot(x,y6,c="purple",label='$O(2^n)$')

y7 = [math.factorial(n) for n in x]
plt.plot(x,y7,c="black",label=("O(n!)"))

y8 = [n**n for n in x]
plt.plot(x,y8,label = "O(n^n)")

plt.title("Big O Notation")
plt.xlabel("Elements")
plt.ylabel("Operation")
plt.xlim(-2,14)
plt.ylim(0,100)
#lets annontate origin, its coo
plt.annotate("origin" , xy = (0,0), xytext = (0,20),
            arrowprops = {'facecolor' :'black',
                         'arrowstyle' : '->'
                          }
            )


plt.legend()
plt.show()
```
